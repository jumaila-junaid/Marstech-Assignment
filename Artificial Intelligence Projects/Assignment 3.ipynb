```python
# print Odd Number in the list
```

```python
numbers = (20,10,16,19,25,1,276,188)
print(numbers)
for num in numbers:
  if num % 2==1:
      print(f"{num} is odd")
```

```python
# print Even numbers in the list
```

```python
numbers = (20,10,16,19,25,1,276,188)
print(numbers)
for num in numbers:
    if num % 2==0:
        print(f"{num} is even")
```
